```python
from dataclasses import dataclass
from pathlib import Path

@dataclass
class ContextualEmbedderConfig:
    # Path to the root directory where embedding artifacts will be stored.
    root_dir: Path
    
    # Path to the file containing results from Semantic Role Labeling (SRL) and possibly other analyses.
    results_path: Path
    
    # Path where the generated embeddings should be saved.
    output_path: Path
    
    # The name of the model to be used from SentenceTransformers for generating embeddings.
    model_name: str
```

# Configuration Manager

```python
from src.career_chief.constants import *
from src.career_chief.utils.common import read_yaml, create_directories
from src.career_chief import logger
from src.career_chief.entity.config_entity import (ContextualEmbedderConfig)

class ConfigurationManager:
    """
    ConfigurationManager manages configurations needed for the data pipeline.

    The class reads configuration, parameter, and schema settings from specified files
    and provides a set of methods to access these settings. It also takes care of
    creating necessary directories defined in the configurations.

    Attributes:
    - config (dict): Configuration settings.
    - params (dict): Parameters for the pipeline.
    - schema (dict): Schema information.
    """
    
    def __init__(self, 
                 config_filepath = CONFIG_FILE_PATH, 
                 params_filepath = PARAMS_FILE_PATH, 
                 schema_filepath = SCHEMA_FILE_PATH) -> None:
        """
        Initialize ConfigurationManager with configurations, parameters, and schema.

        Args:
        - config_filepath (Path): Path to the configuration file.
        - params_filepath (Path): Path to the parameters file.
        - schema_filepath (Path): Path to the schema file.

        Creates:
        - Directories specified in the configuration.
        """
        self.config = self._read_config_file(config_filepath, "config")
        self.params = self._read_config_file(params_filepath, "params")
        self.schema = self._read_config_file(schema_filepath, "schema")

        # Create the directory for storing artifacts if it doesn't exist
        create_directories([self.config.artifacts_root])

    def _read_config_file(self, filepath: str, config_name: str) -> dict:
        """
        Read a configuration file and return its content.

        Args:
        - filepath (str): Path to the configuration file.
        - config_name (str): Name of the configuration (for logging purposes).

        Returns:
        - dict: Configuration settings.

        Raises:
        - Exception: If there's an error reading the file.
        """
        try:
            return read_yaml(filepath)
        except Exception as e:
            logger.error(f"Error reading {config_name} file: {filepath}. Error: {e}")
            raise


    def get_contextual_embeddings_config(self) -> ContextualEmbedderConfig:
        
        try:
            get_contextual_embeddings_config = self.config['contextual_embeddings']
            
            return ContextualEmbedderConfig(
                root_dir=Path(get_contextual_embeddings_config['root_dir']),
                results_path=Path(get_contextual_embeddings_config['results_path']),
                output_path=Path(get_contextual_embeddings_config['output_path']),
                model_name=get_contextual_embeddings_config['model_name']
            )
        except KeyError as e:
            logger.error(f"A required configuration is missing in the 'contextual embeddings config' section: {e}")
            raise KeyError(f"Missing configuration in 'contextual embeddings config': {e}") from e
```

# Component

```python
from sentence_transformers import SentenceTransformer
import pandas as pd
from tqdm import tqdm
import numpy as np
from src.career_chief import logger

class ContextualEmbedder:
    def __init__(self, config):
        """
        Initializes the ContextualEmbedder with the SentenceTransformer model specified in the configuration.
        Loads data from the specified CSV file path provided in the config.
        
        Args:
            config (ConfigClass): Configuration object containing attributes like model_name, results_path, and output_path.
        """
        self.config = config
        self.model = SentenceTransformer(config.model_name)
        self.data = pd.read_csv(config.results_path)
        logger.info(f"ContextualEmbedder initialized with model {config.model_name} and data from {config.results_path}.")

    def preprocess_text(self):
        """
        Processes each row of the dataset to combine various text fields into a single text representation.
        This involves concatenating entity representations, topic information, probability scores, and semantic role labeling results.
        """
        self.data['combined_text'] = self.data.apply(lambda row: ' '.join([
            ' '.join([f"{ent[0]} [{ent[1]}]" for ent in eval(row['entities'])]),  # Assuming entities are stored in a serialized list of tuples
            f"Topic {row['topic']}",
            str(row['probability']),  # Convert float to string explicitly
            row['processed_srl_results']
        ]), axis=1)
        logger.info("Text data preprocessed successfully.")


    def create_embeddings(self):
        """
        Generates embeddings for each entry in the dataframe by encoding the combined text.
        """
        if 'combined_text' not in self.data.columns:
            self.preprocess_text()
        self.data['embeddings'] = [self.model.encode(text, show_progress_bar=True) for text in tqdm(self.data['combined_text'], desc="Encoding")]
        logger.info("Embeddings created successfully.")

    def save_embeddings(self):
        """
        Saves the embeddings to the specified output path in the configuration.
        """
        output_path = self.config.output_path
        self.data[['job_id', 'embeddings']].to_pickle(output_path)
        logger.info(f"Embeddings saved successfully at {output_path}.")

    def get_embeddings(self):
        """
        Returns the embeddings data.
        """
        return self.data['embeddings']
```

# Pipeline

```python
from src.career_chief.config.configuration import ConfigurationManager
# from src.career_chief.components.contextual_embedding import ContextualEmbedder  # Ensure this import matches the actual path and class name
from src.career_chief import logger

class ContextualEmbedderPipeline:
    """
    A pipeline to process text data, create embeddings, and save them locally.
    """

    STAGE_NAME = "Contextual Embedding Pipeline"

    def __init__(self):
        """
        Initializes the pipeline with the configuration manager to obtain settings.
        """
        self.config_manager = ConfigurationManager()
        logger.info(f"{self.STAGE_NAME} initialized successfully.")

    def run(self):
        """
        Executes the pipeline to preprocess text, create embeddings, and save them locally.
        """
        try:
            logger.info(f"{self.STAGE_NAME}: Fetching configuration for Contextual Embedder.")
            embeddings_config = self.config_manager.get_contextual_embeddings_config()

            logger.info(f"{self.STAGE_NAME}: Initializing the Contextual Embedder.")
            contextual_embedder = ContextualEmbedder(embeddings_config)

            logger.info(f"{self.STAGE_NAME}: Starting preprocessing and embedding creation.")
            contextual_embedder.create_embeddings()

            logger.info(f"{self.STAGE_NAME}: Saving embeddings.")
            contextual_embedder.save_embeddings()

            logger.info(f"{self.STAGE_NAME}: Embedding process completed successfully.")
        
        except Exception as e:
            logger.error(f"{self.STAGE_NAME}: Error occurred - {str(e)}")
            raise e

if __name__ == '__main__':
    pipeline = ContextualEmbedderPipeline()
    pipeline.run()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

def load_embeddings(file_path):
    """
    Loads embeddings from a pickle file.
    
    Args:
        file_path (str): Path to the pickle file containing the embeddings.
        
    Returns:
        pd.DataFrame: DataFrame containing the embeddings.
    """
    return pd.read_pickle(file_path)

def visualize_embeddings(embeddings):
    """
    Visualizes embeddings using t-SNE.
    
    Args:
        embeddings (np.array): Numpy array of embeddings.
    """
    tsne = TSNE(n_components=2, perplexity=2, n_iter=1000)
    reduced_data = tsne.fit_transform(embeddings)

    plt.figure(figsize=(8, 5))
    plt.scatter(reduced_data[:, 0], reduced_data[:, 1])
    plt.title('t-SNE Visualization of Job Embeddings')
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.grid(True)
    plt.show()

# Path to your embeddings file
file_path = 'artifacts/model_training/context_embeddings/output/jobs_embeddings.pkl'
embeddings_df = load_embeddings(file_path)

# Assuming 'embeddings' column is a list of lists
embeddings_array = np.stack(embeddings_df['embeddings'].values)
visualize_embeddings(embeddings_array)
```

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns

def visualize_embeddings(file_path):
    """
    Loads embeddings from a pickle file, reduces their dimensionality, and visualizes them in a 2D plot.

    Args:
        file_path (str): Path to the pickle file containing the embeddings.
    """
    # Load the embeddings
    embeddings = pd.read_pickle(file_path)
    embeddings_matrix = np.stack(embeddings['embeddings'].values)

    # Reduce dimensions with PCA to reduce computational cost for t-SNE and noise
    pca = PCA()
    pca_result = pca.fit_transform(embeddings_matrix)
    print(f'Cumulative explained variation for 50 principal components: {np.sum(pca.explained_variance_ratio_)}')

    # Further reduce the dimensions with t-SNE
    tsne = TSNE(n_components=2, perplexity=30, n_iter=300)
    tsne_results = tsne.fit_transform(pca_result)

    # Create a plot
    plt.figure(figsize=(12, 8))
    sns.scatterplot(x=tsne_results[:,0], y=tsne_results[:,1], palette=sns.color_palette("hsv", 10), legend="full", alpha=0.5)

    plt.title('t-SNE Visualization of Job Embeddings')
    plt.xlabel('t-SNE Component 1')
    plt.ylabel('t-SNE Component 2')
    plt.show()

# Specify the path to your embeddings file
file_path = 'artifacts/model_training/context_embeddings/output/jobs_embeddings.pkl'
visualize_embeddings(file_path)
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

def load_and_merge_data(embeddings_path, job_info_path):
    """
    Loads embeddings and job information, merges them on job_id.

    Args:
        embeddings_path (str): Path to the pickle file with embeddings.
        job_info_path (str): Path to the CSV file with job information.

    Returns:
        pd.DataFrame: Merged DataFrame with embeddings and job titles.
    """
    embeddings_df = pd.read_pickle(embeddings_path)
    job_info_df = pd.read_csv(job_info_path)
    # Make sure 'job_id' is the common column in both DataFrames
    merged_df = pd.merge(embeddings_df, job_info_df, on='job_id')
    return merged_df

def plot_embeddings_with_labels(df, label_column='job_title', num_labels=10):
    """
    Performs PCA on embeddings and plots the first two principal components.
    Labels a subset of points with job titles for better visualization.

    Args:
        df (pd.DataFrame): DataFrame containing 'embeddings' and job titles.
        label_column (str): Column name of the labels to display on the plot.
        num_labels (int): Number of points to label on the plot.
    """
    # Convert embeddings from list to numpy array
    embeddings_matrix = np.stack(df['embeddings'].values)
    pca = PCA(n_components=2)
    reduced_embeddings = pca.fit_transform(embeddings_matrix)

    # Plotting
    plt.figure(figsize=(12, 8))
    plt.scatter(reduced_embeddings[:, 0], reduced_embeddings[:, 1], alpha=0.6, edgecolors='w', s=100)
    
    # Annotate points
    for i, label in enumerate(df[label_column]):
        if i < num_labels:  # Label the first num_labels points
            plt.annotate(label, (reduced_embeddings[i, 0], reduced_embeddings[i, 1]), fontsize=9, ha='right')

    plt.title('PCA of Job Embeddings with Job Titles')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.grid(True)
    plt.show()

# Specify paths
embeddings_path = 'artifacts/model_training/context_embeddings/output/jobs_embeddings.pkl'
job_info_path = 'artifacts/model_training/allennlp-models/output/srl_results.csv'

# Load and merge data
merged_data = load_and_merge_data(embeddings_path, job_info_path)

# Plot with labels
plot_embeddings_with_labels(merged_data, 'title', num_labels=10)
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load embeddings
embeddings_path = 'artifacts/model_training/context_embeddings/output/jobs_embeddings.pkl'
embeddings_df = pd.read_pickle(embeddings_path)

# Load job information
job_info_path = 'artifacts/model_training/allennlp-models/output/srl_results.csv'
job_info_df = pd.read_csv(job_info_path)

# Merge the dataframes on job_id if necessary
# Assuming both dataframes have a 'job_id' column
full_df = pd.merge(job_info_df, embeddings_df, on='job_id')

# Define the plot functions
def plot_clustered_embeddings(df, n_clusters=5, num_labels=10):
    embeddings_matrix = np.stack(df['embeddings'].values)
    pca = PCA(n_components=2)
    reduced_embeddings = pca.fit_transform(embeddings_matrix)
    
    kmeans = KMeans(n_clusters=n_clusters)
    labels = kmeans.fit_predict(reduced_embeddings)

    plt.figure(figsize=(12, 8))
    scatter = plt.scatter(reduced_embeddings[:, 0], reduced_embeddings[:, 1], c=labels, cmap='viridis', alpha=0.6, edgecolors='w', s=100)
    plt.colorbar(scatter, label='Cluster')

    for i, txt in enumerate(df['title']):
        if i < num_labels:
            plt.annotate(txt, (reduced_embeddings[i, 0], reduced_embeddings[i, 1]), fontsize=9, ha='right')

    plt.title('PCA of Job Embeddings with K-means Clustering')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.grid(True)
    plt.show()

def plot_similarity_heatmap(df):
    embeddings_matrix = np.stack(df['embeddings'].values)
    similarity_matrix = cosine_similarity(embeddings_matrix)

    plt.figure(figsize=(12, 10))
    sns.heatmap(similarity_matrix, cmap='coolwarm', xticklabels=False, yticklabels=False)
    plt.title('Heatmap of Cosine Similarity Among Jobs')
    plt.xlabel('Jobs')
    plt.ylabel('Jobs')
    plt.show()

# Run plotting functions
plot_clustered_embeddings(full_df, n_clusters=5, num_labels=10)
plot_similarity_heatmap(full_df)
```

```python
import pandas as pd

def save_sampled_data(file_path, output_path, fraction=0.1, random_state=42):
    """
    Saves a specified fraction of the dataset to a CSV file.

    Args:
        file_path (str): The path to the original full dataset in CSV format.
        output_path (str): The path where the sampled dataset will be saved.
        fraction (float): The fraction of the data to sample and save. Default is 0.3 (30%).
        random_state (int): A seed value to ensure reproducibility. Default is 42.

    Returns:
        None
    """
    # Load the full dataset
    data = pd.read_csv(file_path)
    
    # Sample the dataset
    sampled_data = data.sample(frac=fraction, random_state=random_state)
    
    # Save the sampled dataset to a new CSV file
    sampled_data.to_csv(output_path, index=False)
    
    print(f"Sampled data saved to {output_path}. Total records saved: {len(sampled_data)}")

# Example usage:
file_path = 'artifacts/model_training/context_embeddings/output/merged.csv'
output_path = 'artifacts/model_training/context_embeddings/output/merged_sample.csv'
save_sampled_data(file_path, output_path)
```
